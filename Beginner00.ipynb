```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.read_csv("Salary_dataset.csv")
```

```python
#Removendo da tabela uma base n√£o utilizada, para evitar que os dados irrelevantes impactem no resultado final do modelo
#O axis=1 serve para referenciar que estamos procurando por uma coluna, se fosse igual a 0, estariamos referenciando uma fileira
data = data.drop(["Unnamed: 0"],axis=1)
data.head()
```

```python
# Configurando o estilo
sns.set_theme(palette=["#9b91c0"])

# Plot data points com seaborn
#O m√©todo joinplot √© respons√°vel por criar os gr√°fico
sns.jointplot(data=data, x="YearsExperience", y="Salary")
```

```python
# Separate the features (YearsExperience) from the target variable (Salary)
X = data.YearsExperience
Y = data.Salary

#O algoritmo espera uma matriz com duas dimens√µes e meus dados possuem apenas uma dimens√£o, por isso √© necess√°rio remodelar
#Um array   1D √© assim: [1, 2, 3, 4, 5]
#Um array 2d √© como uma tabela, assim:
#[
#  [1, 2, 3],
#  [4, 5, 6],
#  [7, 8, 9]
#]

X = X.values.reshape(-1, 1)
Y = Y.values.reshape(-1, 1)
```

```python
# Create a linear regression model
regression = LinearRegression()

# Treinando o modelo
regression.fit(X, Y)

# Predict the Salary values based on YearsExperience
#predictions = regression.predict(X)
```

```python
# Criar um plot de regress√£o linear
plot = sns.lmplot(data=data, x='YearsExperience', y='Salary', aspect=1.5, height=6)

# Labels
plot.set_axis_labels('Years of Experience', 'Salary')

#Titulo
plt.title('Linear Regression: Years of Experience vs Salary')

# Mostrar o plor
plt.show()
```

```python
# Organizar os dados de entrada em um array numpy
year_of_experience = [[15]]

# Fazer a previs√£o
previsao = regression.predict(year_of_experience)

# Exibir a previs√£o
print("Predicted salary for 15 years of experience : ", previsao[0][0])
```
